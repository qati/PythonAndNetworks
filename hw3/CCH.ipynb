```python
# import zipfile for opening the zip
import zipfile
#import numpy
import numpy as np
```

```python
# open the zip, from opened zip open the text file
zf = zipfile.ZipFile("Atlantic16.zip")
f  = zf.open("Atlantic16.txt", "r")
```

```python
#read from every line the price of property and store in prices list
prices = []
for line in f:
    p = int( (str(line)[2:-5])[439:447] )
    prices.append(p)
```

```python
# sorting the prices array, heapsort nlog(n) algorithm
prices = np.sort(np.array(prices), kind="heapsort")
```

```python
def cch(x, arr):
    # finding the x value in the prices array
    # searchsorted use Binary search, log(n) algorihm
    idx = np.searchsorted(arr, int(np.ceil(x)))
    return len(arr)-idx-1
```

```python
# creating x values between 0 and 1e6, calculating cch for all x
x = [i for i in range(1000000)]
h = [cch(i, prices) for i in x]
```

```python
# import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#plot
plt.figure(0, (16,8))
plt.xlabel("Price", fontsize=16)
plt.ylabel("CCH(Price)", fontsize=16)
plt.plot(x, h)
```
